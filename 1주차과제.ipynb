* **네이버의 파파고**

* **네이버 클라우드 플랫폼의 챗봇(Chatbot)**

    네이버의 노하우가 축척된 자연어 처리 기술로 사용자의 질문 
 
    의도를 정확하게 파악하고 자연스러운 대화를 이어간다.
 
    딥 러닝 기반 자가 학습 알고리즘 적용으로 발화 예시
 
    입력 횟수를 줄이고 대화 모델을 최적화할 수 있다.
 

* **솔트룩스의 자연어이해 엔진 LEA**
 
    
    자연어이해 엔진(LEA-Language Engineering & Analysis)은 비정형 데이터
    
    가공을 위해 형태소 분석, 개체명 인식, 구문 분석, 감성 분석 
    
    등의 텍스트분석 기능을 처리하는 기계학습/심층학습 기반의 언어 분석 엔진이다.
    
    자연어이해 엔진을 구성하고 있는 고정밀 언어분석기들은 기계학습과
    
    심층학습(인공신경망) 기술이 적용되어 있으며, 한국어뿐 아니라 다국어
    
    대응이 가능하며 지식그래프와 연계하여 의미해석,질의응답,대화 시스템
    
    구현이 가능한 자연어처리 엔진이다.
 
 

* **솔트룩스의 심층QA 엔진**
 
    
    사용자의 질문에 대해 다양한 지식을 스스로 학습, 추론하여 축적한
 
    지식으로부터 최적의 답을 찾아내어 답을 제시하는 시스템이다.
 
    지식베이스 기반 질의응답시룰(KBQA),정보 검색 기반 질의응답기술(IRQA),
 
    기계독해 기술을 이용한 질의응답기술(MRCQA), 상담 및 대화 이력학습
 
    기반 질의응답기술(DLQA)등이 앙상블 되어 질의 유형에 따라 최적의 풀이
 
    방식을 채택하여 사용자에게 응답을 제시합니다.

* **구글의 Siri**

    사용자가 말을 하면 그것을 녹음해 애플의 서버로 전송한 다음,

    그러면 그 내용을 애플이 인수한 SRI의 인공지능으로 분석하여

* **아마존 에코**

* **SK텔레콤의 NUAU**

> **인공지능의 이미지기술 및 서비스 분야**
    
    
    이미지 인식 기술은 이미지의 정보를 식별하는 기술로서 컴퓨터 비전
 
    분야에서 가장 오랫동안 연구된 기술 중 하나이다. 오랜 역사만큼 기술
 
    의 성숙도도 높은 편이라 보안 감시, 의료, 자율주행 등의 산업 분야
 
    뿐만 아니라, 헬스케어, 스마트 홈, 엔터테인먼트 등의 우리 주변 실생활
 
    곳곳에 많이 활용되고 있다.
 
* **OCR(optical character recognition)**
 
    
    OCR은 한국어로 풀자면 광학식 문자 판독이라고한다.
 
    보통 컴퓨터가 2진법 데이터를 폰트를 통해 인간이 인식할 수 있는 형태로
 
    글자를 보여 준다면, OCR은 그 반대로 인간이 종이 위에 써 놓은 글씨를
 
    인지하여 텍스트 데이터로 치환한다.
 
    현대의 OCR은 인쇄물 상의 글자와 이미지를 디지털 데이터로 변해주는
 
    자동인식 기술이다. 명함이나 서류,서적 등의 정보를 빠르고 간편하게
 
    변환하여 활용 할수 있다.
 
* **구글 드라이브**
 
    
    다른 웹하드보다 용량은 작지만 뷰어 기능이 상당히 강력하다.
 
    HD 동영상, Adobe lllustrator, PNG, Photoshop 등 30여종의 파일을
 
    열수 있고 이미지에 포함된 개체와 스캔한 문서에 있는 텍스트도 인식할
 
    수 있다. 예를 들어 캘리포니아를 검색하면 드라이브 안에 있는 캘리
 
    포니아 사진을 로딩하며 신문 기사의 내용을 검색하면 스캔한 신문
 
    기사의 파일을 인식해 보여준다.
 
* **페이스북의 딥 페이스**
 
    
    딥 페이스라 불리는 얼굴 인식 알고리즘은 페이스 북
 
    에 사진을 올리면 알고리즘이 알아서 사진 속 인물의
 
    이름을 알아내며 정확도가 97.25%에 달한다. 이러한
 
    원리는 주로 두 눈 사이의 거리, 코의 길이와 너비, 턱
 
    선의 길이와 같이 수로 나타낼 수 있는 특징이 그 비교
 
    대상이 된다.
 
* **구글 포토**
 
    
    구글포토는 머신러닝을 이용해서 장소 및 사물을 인식
 
    하고, 사용자가 태그를 지정하지 않아도 알아서 분류를
 
    해놓는다.특히 장소를 찾는 알고리즘은 수단과 방법을 가리지
 
    않고 모든 데이터를 다 수집해서 찾아준다. 기본적으로 사진
 
    안에 들어있는 GPS정보를 이용하고, 만약 사진에 GPS정보가 없
 
    어도 사진을 분석해서 랜드마크를 인식하는 방법으로 위치를
 
    찾아낸다.
    
> **인공지능의 자율주행기술 및 서비스 분야**
 
    
    자율주행은 무생물인 교통수단이 사람없이 스스로 판단하고
 
    운행하는 것을 뜻한다. 무인운전과 혼용되어 쓰이긴 하지만
 
    무인운전은 사람이 없지만 외부에 통제를 받아 주행하는것 
 
    이지만 자율주행은 사람에 의해 통제받지 않고 탑재된
 
    인공지능 컴퓨터가 스스로 판단하여 주행하는 것이다. 
 
* **테슬라 오토파일럿**
 
    
    테슬라 자동차에서 사용되고 있는 ADAS(첨단 운전자 보조 시스템)
 
    으로서, 아직은 소프트웨어가 완성되지 않아 완전자율주행이
 
    가능하지 않지만 테슬라 소프트웨어 업데이트를 통해서 기능
 
    들이 국가의 법에 맞는것이 확인될때 완전 자율주행이 가능할
 
    예정이다.
 
* **자동운항**
 
    
    비행기가 개발되고 난 직후에는 안전하게 날아디니기 위해 
 
    조종사가 항상 신경쓰고 있었어야 했는데, 이는 비행시간이
 
    늘어날수록 심각한 피로의 누적을 가져오게 되었다. 이에
 
    1919년 자이로스코프를 이용하여 조종사가 진로에 신경쓰지
 
    않고도 똑바로 날 수 있도록 하는 오토파일럿이 최초 개발되었
 
    다.
 
* **열차자동운전장치**
 
    
    열차자동운전장치[Autumatic Train Operation(ATO)]은
 
    열차 속도를 제어하게 되므로, 과속으로 달리면 신호를
 
    받고 열차의 속도를 줄일 수도 있게 되었다. 그렇다면
 
    "사실상 자동운전이 가능하지 않을까"라는 생각으로
 
    만들어낸 시스템이 바로 ATO이다.
    
    S-ATO는 반자동으로서, 승무원이 출입문을 닫은후 취급을
 
    하면 다음 역에서 출입문이 열리기 까지의 모든 과정이 자동
 
    으로 진행된다.
 
* **자율주행 서비스로봇**
 
    
    국내 네이버랩스-현대중공업 등에서 개발하고 있으며,
 
    서비스로봇은 크게 전문 로봇과 개인 로봇이라는 두 가지 종류
 
    로 제조현장을 넘어 우리의 생활과 삶을 스마트하게 혁신하고,
 
    스스로 상황을 판단하여 자율적으로 동작과 반응하는 로봇으로
 
    이용되며 약 258조 5천억원의 시장이 형성될것으로 전망하고
 
    있다. 네이버랩스가 개발한 M1과 어라운드 로봇은 실내공간을
 
    촬영, 정밀 지도를 클라우드에 저장하고, 이를 기반으로 어라운
 
    드가 자율주행하는 시스템을 갖추고 있다.
